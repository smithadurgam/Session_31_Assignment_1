```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

```python
centers = [[1, 1], [-1, -1], [1, -1]]
iris = datasets.load_iris()

X = iris.data
y = iris.target

fig = plt.figure(1, figsize=(14, 11))

ax = Axes3D(fig, rect=[2, 2, 1, 1], elev=70, azim=100)

plt.cla()
pcadec = decomposition.PCA(n_components=3)
pcadec.fit(X)
X = pcadec.transform(X)

# Color labels as assignment specified, and scale dots by value
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.Set1, edgecolor='k', sizes=(20, 200))

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
```
